# рҹҸ—пёҸ SOFTWARE ARCHITECTURE PATTERNS MASTERY

## рҹҺҜ Architect Systems Like a Staff/Principal Engineer

This notebook is a pattern decision guide: when to use each architecture style, when not to, and what trade-offs to accept.

### Pattern Families Covered
- **Microservices** for domain-driven scaling and team autonomy
- **Event-driven systems** for loose coupling and async workflows
- **Clean/Hexagonal architecture** for maintainable codebases
- **CQRS + event sourcing** for high-audit or read/write separation needs
- **Serverless patterns** for bursty workloads and reduced ops overhead
- **Resilience patterns** (circuit breaker, bulkhead, retries, saga)

### Decision Framework (Use Per Problem)
1. Identify business constraints (latency, consistency, release velocity).
2. Choose architecture style that fits those constraints.
3. Define service boundaries and integration model.
4. Plan failure handling and observability from day one.
5. Include migration path from current to target architecture.

### Interview Delivery Template
```
1) Requirements and constraints         8-10 min
2) Candidate architecture options       8-10 min
3) Chosen architecture + rationale     15-20 min
4) Risks, bottlenecks, mitigations     10-12 min
5) Evolution and rollout strategy       5-8 min
```

### Target Outcomes
- Select patterns based on constraints, not trends.
- Explain architecture trade-offs clearly to stakeholders.
- Propose practical migration plans for real teams.

---

**Outcome:** You should be able to justify architecture choices under interview pressure and in production planning discussions.

## Chapter 1: Microservices Architecture вӯҗвӯҗвӯҗ
> Design service-oriented systems that can scale teams and traffic safely.

### Chapter Objectives
- Choose service boundaries based on business domains.
- Prevent cascading failures using resilience patterns.
- Handle cross-service data and transaction consistency.

### Core Principles
- One service, one business capability
- Independent deployment and scaling
- Fault isolation by default
- Data ownership per service
- Clear API contracts and observability

### Implementation Patterns to Apply
- Service discovery and health-aware routing
- API gateway and authentication edge controls
- Circuit breaker + retry + timeout policies
- Saga and outbox patterns for consistency
- Event-driven integration where async is acceptable

````python
# ARCHITECTURE PATTERNS - DESIGN LIKE A PRINCIPAL ENGINEER!

print("рҹҺҜ SOFTWARE ARCHITECTURE PATTERNS - SYSTEMS THAT SCALE!")
print("=" * 75)

from abc import ABC, abstractmethod
from typing import Dict, List, Any, Optional, Callable
from enum import Enum
from dataclasses import dataclass
import json
import time
from datetime import datetime

# ===============================================================
# 1. MICROSERVICES ARCHITECTURE - Netflix Scale
# ===============================================================
print("\nрҹ”Ҙ 1. MICROSERVICES ARCHITECTURE")
print("Build systems like Netflix, Uber, and Amazon")

class ServiceHealth(Enum):
    HEALTHY = "healthy"
    UNHEALTHY = "unhealthy"
    DEGRADED = "degraded"

@dataclass
class ServiceInstance:
    service_name: str
    host: str
    port: int
    health: ServiceHealth
    version: str

class ServiceRegistry:
    """Service Discovery - Find and communicate with services"""
    
    def __init__(self):
        self._services: Dict[str, List[ServiceInstance]] = {}
    
    def register(self, instance: ServiceInstance):
        """Register a service instance"""
        if instance.service_name not in self._services:
            self._services[instance.service_name] = []
        self._services[instance.service_name].append(instance)
        print(f"  вң… Registered {instance.service_name} at {instance.host}:{instance.port}")
    
    def discover(self, service_name: str) -> List[ServiceInstance]:
        """Discover healthy instances of a service"""
        instances = self._services.get(service_name, [])
        healthy_instances = [i for i in instances if i.health == ServiceHealth.HEALTHY]
        return healthy_instances
    
    def health_check(self, service_name: str, host: str, port: int) -> bool:
        """Simple health check simulation"""
        # In real world, this would be HTTP health check
        import random
        return random.random() > 0.1  # 90% success rate

class APIGateway:
    """Single entry point for all client requests"""
    
    def __init__(self, service_registry: ServiceRegistry):
        self.service_registry = service_registry
        self.rate_limits: Dict[str, int] = {}
        self.circuit_breakers: Dict[str, 'CircuitBreaker'] = {}
    
    def route_request(self, service_name: str, endpoint: str, payload: Dict) -> Dict:
        """Route request to appropriate microservice"""
        
        # Rate limiting
        if not self._check_rate_limit(service_name):
            return {"error": "Rate limit exceeded", "status": 429}
        
        # Service discovery
        instances = self.service_registry.discover(service_name)
        if not instances:
            return {"error": f"Service {service_name} not available", "status": 503}
        
        # Load balancing (simple round-robin)
        instance = instances[0]  # In real world, use proper load balancing
        
        # Circuit breaker check
        circuit_breaker = self._get_circuit_breaker(service_name)
        if circuit_breaker.is_open():
            return {"error": f"Circuit breaker open for {service_name}", "status": 503}
        
        try:
            # Simulate service call
            response = self._call_service(instance, endpoint, payload)
            circuit_breaker.record_success()
            return response
        except Exception as e:
            circuit_breaker.record_failure()
            return {"error": str(e), "status": 500}
    
    def _check_rate_limit(self, service_name: str) -> bool:
        """Simple rate limiting"""
        # In real world, use Redis or similar for distributed rate limiting
        current_requests = self.rate_limits.get(service_name, 0)
        if current_requests < 100:  # 100 requests per time window
            self.rate_limits[service_name] = current_requests + 1
            return True
        return False
    
    def _get_circuit_breaker(self, service_name: str) -> 'CircuitBreaker':
        """Get or create circuit breaker for service"""
        if service_name not in self.circuit_breakers:
            self.circuit_breakers[service_name] = CircuitBreaker(
                failure_threshold=5,
                timeout=30
            )
        return self.circuit_breakers[service_name]
    
    def _call_service(self, instance: ServiceInstance, endpoint: str, payload: Dict) -> Dict:
        """Simulate calling a microservice"""
        # In real world, this would be HTTP request
        if instance.health != ServiceHealth.HEALTHY:
            raise Exception(f"Service {instance.service_name} is unhealthy")
        
        return {
            "status": 200,
            "data": f"Response from {instance.service_name}:{endpoint}",
            "service_version": instance.version
        }

class CircuitBreaker:
    """Prevent cascade failures between services"""
    
    def __init__(self, failure_threshold: int = 5, timeout: int = 60):
        self.failure_threshold = failure_threshold
        self.timeout = timeout
        self.failure_count = 0
        self.last_failure_time = None
        self.state = "CLOSED"  # CLOSED, OPEN, HALF_OPEN
    
    def is_open(self) -> bool:
        """Check if circuit breaker is open"""
        if self.state == "OPEN":
            if time.time() - self.last_failure_time > self.timeout:
                self.state = "HALF_OPEN"
                return False
            return True
        return False
    
    def record_success(self):
        """Record successful operation"""
        self.failure_count = 0
        self.state = "CLOSED"
    
    def record_failure(self):
        """Record failed operation"""
        self.failure_count += 1
        self.last_failure_time = time.time()
        
        if self.failure_count >= self.failure_threshold:
            self.state = "OPEN"
            print(f"  рҹҡЁ Circuit breaker OPENED (failures: {self.failure_count})")

# Demonstrate microservices architecture
print("\nрҹ“Ҡ MICROSERVICES IN ACTION:")

# Set up service registry
registry = ServiceRegistry()

# Register microservices
services = [
    ServiceInstance("user-service", "10.0.1.10", 8080, ServiceHealth.HEALTHY, "v1.2.0"),
    ServiceInstance("user-service", "10.0.1.11", 8080, ServiceHealth.HEALTHY, "v1.2.0"),
    ServiceInstance("order-service", "10.0.2.10", 8081, ServiceHealth.HEALTHY, "v2.1.0"),
    ServiceInstance("payment-service", "10.0.3.10", 8082, ServiceHealth.HEALTHY, "v1.0.5"),
    ServiceInstance("notification-service", "10.0.4.10", 8083, ServiceHealth.HEALTHY, "v1.1.0")
]

for service in services:
    registry.register(service)

# Set up API Gateway
gateway = APIGateway(registry)

# Simulate client requests
requests = [
    ("user-service", "/users/123", {"action": "get_profile"}),
    ("order-service", "/orders", {"user_id": 123, "items": ["item1", "item2"]}),
    ("payment-service", "/payments", {"order_id": 456, "amount": 99.99}),
    ("notification-service", "/notify", {"user_id": 123, "message": "Order confirmed"})
]

print("\n  Client Requests:")
for service, endpoint, payload in requests:
    response = gateway.route_request(service, endpoint, payload)
    print(f"    {service}{endpoint}: {response}")

# ===============================================================
# 2. EVENT-DRIVEN ARCHITECTURE - Uber Scale
# ===============================================================
print("\n\nрҹ”Ҙ 2. EVENT-DRIVEN ARCHITECTURE")
print("Build reactive systems like Uber and Spotify")

@dataclass
class Event:
    event_type: str
    data: Dict[str, Any]
    timestamp: datetime
    source_service: str
    event_id: str

class EventBus:
    """Central event distribution system"""
    
    def __init__(self):
        self._subscribers: Dict[str, List[Callable]] = {}
        self._event_store: List[Event] = []
    
    def subscribe(self, event_type: str, handler: Callable):
        """Subscribe to specific event types"""
        if event_type not in self._subscribers:
            self._subscribers[event_type] = []
        self._subscribers[event_type].append(handler)
        print(f"  рҹ“Ў Subscribed to {event_type} events")
    
    def publish(self, event: Event):
        """Publish event to all subscribers"""
        self._event_store.append(event)
        print(f"  рҹ“Ө Published: {event.event_type} from {event.source_service}")
        
        # Notify subscribers
        subscribers = self._subscribers.get(event.event_type, [])
        for handler in subscribers:
            try:
                handler(event)
            except Exception as e:
                print(f"    вқҢ Handler error: {e}")
    
    def get_events(self, event_type: Optional[str] = None) -> List[Event]:
        """Get events from event store"""
        if event_type:
            return [e for e in self._event_store if e.event_type == event_type]
        return self._event_store.copy()

class EventDrivenService(ABC):
    """Base class for event-driven microservices"""
    
    def __init__(self, service_name: str, event_bus: EventBus):
        self.service_name = service_name
        self.event_bus = event_bus
        self._setup_event_handlers()
    
    @abstractmethod
    def _setup_event_handlers(self):
        """Set up event subscriptions"""
        pass
    
    def emit_event(self, event_type: str, data: Dict[str, Any]):
        """Emit an event"""
        event = Event(
            event_type=event_type,
            data=data,
            timestamp=datetime.now(),
            source_service=self.service_name,
            event_id=f"{self.service_name}_{int(time.time())}"
        )
        self.event_bus.publish(event)

class OrderService(EventDrivenService):
    """Order management service"""
    
    def _setup_event_handlers(self):
        self.event_bus.subscribe("user.registered", self._handle_user_registered)
        self.event_bus.subscribe("payment.processed", self._handle_payment_processed)
    
    def create_order(self, user_id: int, items: List[str], total: float):
        """Create a new order"""
        order_id = f"order_{int(time.time())}"
        
        # Business logic
        print(f"    рҹӣ’ Creating order {order_id} for user {user_id}")
        
        # Emit event
        self.emit_event("order.created", {
            "order_id": order_id,
            "user_id": user_id,
            "items": items,
            "total": total
        })
        
        return order_id
    
    def _handle_user_registered(self, event: Event):
        """Handle user registration event"""
        user_id = event.data.get("user_id")
        print(f"    рҹ‘Ӣ Order service: Welcome user {user_id}! Here's a 10% discount.")
    
    def _handle_payment_processed(self, event: Event):
        """Handle payment processed event"""
        order_id = event.data.get("order_id")
        print(f"    вң… Order service: Order {order_id} payment confirmed, preparing shipment")

class PaymentService(EventDrivenService):
    """Payment processing service"""
    
    def _setup_event_handlers(self):
        self.event_bus.subscribe("order.created", self._handle_order_created)
    
    def _handle_order_created(self, event: Event):
        """Handle order creation event"""
        order_id = event.data.get("order_id")
        total = event.data.get("total")
        
        print(f"    рҹ’і Payment service: Processing payment for order {order_id}")
        
        # Simulate payment processing
        if total < 1000:  # Assume payments under $1000 always succeed
            self.emit_event("payment.processed", {
                "order_id": order_id,
                "amount": total,
                "status": "success"
            })
        else:
            self.emit_event("payment.failed", {
                "order_id": order_id,
                "amount": total,
                "reason": "Amount too high"
            })

class NotificationService(EventDrivenService):
    """Notification service"""
    
    def _setup_event_handlers(self):
        self.event_bus.subscribe("order.created", self._handle_order_created)
        self.event_bus.subscribe("payment.processed", self._handle_payment_processed)
        self.event_bus.subscribe("payment.failed", self._handle_payment_failed)
    
    def _handle_order_created(self, event: Event):
        """Handle order creation"""
        user_id = event.data.get("user_id")
        order_id = event.data.get("order_id")
        print(f"    рҹ“§ Notification: Order {order_id} confirmation sent to user {user_id}")
    
    def _handle_payment_processed(self, event: Event):
        """Handle successful payment"""
        order_id = event.data.get("order_id")
        print(f"    рҹ“§ Notification: Payment confirmation for order {order_id}")
    
    def _handle_payment_failed(self, event: Event):
        """Handle failed payment"""
        order_id = event.data.get("order_id")
        print(f"    рҹ“§ Notification: Payment failed for order {order_id}")

# Demonstrate event-driven architecture
print("\nрҹ“Ҡ EVENT-DRIVEN SYSTEM IN ACTION:")

# Set up event bus and services
event_bus = EventBus()
order_service = OrderService("order-service", event_bus)
payment_service = PaymentService("payment-service", event_bus)
notification_service = NotificationService("notification-service", event_bus)

# Simulate user registration (external event)
print("\n  User Registration Flow:")
user_event = Event(
    event_type="user.registered",
    data={"user_id": 12345, "email": "user@example.com"},
    timestamp=datetime.now(),
    source_service="user-service",
    event_id="user_reg_001"
)
event_bus.publish(user_event)

# Simulate order creation
print("\n  Order Creation Flow:")
order_id = order_service.create_order(
    user_id=12345,
    items=["laptop", "mouse"],
    total=899.99
)

print(f"\n  рҹ“Ҳ Total events in system: {len(event_bus.get_events())}")

# ===============================================================
# 3. CLEAN ARCHITECTURE - Uncle Bob's Principles
# ===============================================================
print("\n\nрҹ”Ҙ 3. CLEAN ARCHITECTURE")
print("Organize code like Google and maintain it for decades")

# Domain Layer - Business Rules
class User:
    """Domain entity - pure business logic"""
    def __init__(self, user_id: str, email: str, username: str):
        self.user_id = user_id
        self.email = email
        self.username = username
        self.is_active = True
    
    def deactivate(self):
        """Business rule: deactivate user"""
        self.is_active = False
    
    def change_email(self, new_email: str):
        """Business rule: validate and change email"""
        if "@" not in new_email:
            raise ValueError("Invalid email format")
        self.email = new_email

# Use Case Layer - Application Business Rules
class UserRepository(ABC):
    """Port - interface for data access"""
    @abstractmethod
    def save(self, user: User) -> bool:
        pass
    
    @abstractmethod
    def find_by_id(self, user_id: str) -> Optional[User]:
        pass

class CreateUserUseCase:
    """Use case - application-specific business rules"""
    def __init__(self, user_repository: UserRepository):
        self.user_repository = user_repository
    
    def execute(self, email: str, username: str) -> str:
        """Create a new user"""
        # Business logic
        if not email or not username:
            raise ValueError("Email and username are required")
        
        user_id = f"user_{int(time.time())}"
        user = User(user_id, email, username)
        
        # Save using repository
        success = self.user_repository.save(user)
        if not success:
            raise Exception("Failed to create user")
        
        return user_id

# Interface Adapters Layer - Convert data formats
class InMemoryUserRepository(UserRepository):
    """Adapter - concrete implementation of repository"""
    def __init__(self):
        self._users: Dict[str, User] = {}
    
    def save(self, user: User) -> bool:
        """Save user to in-memory storage"""
        self._users[user.user_id] = user
        print(f"    рҹ’ҫ Saved user {user.user_id} to database")
        return True
    
    def find_by_id(self, user_id: str) -> Optional[User]:
        """Find user by ID"""
        return self._users.get(user_id)

class UserController:
    """Controller - handle HTTP requests"""
    def __init__(self, create_user_use_case: CreateUserUseCase):
        self.create_user_use_case = create_user_use_case
    
    def create_user_endpoint(self, request_data: Dict) -> Dict:
        """HTTP endpoint for user creation"""
        try:
            email = request_data.get("email")
            username = request_data.get("username")
            
            user_id = self.create_user_use_case.execute(email, username)
            
            return {
                "status": "success",
                "user_id": user_id,
                "message": "User created successfully"
            }
        except Exception as e:
            return {
                "status": "error",
                "message": str(e)
            }

# Demonstrate clean architecture
print("\nрҹ“Ҡ CLEAN ARCHITECTURE IN ACTION:")

# Dependency injection - outer layers depend on inner layers
repository = InMemoryUserRepository()
use_case = CreateUserUseCase(repository)
controller = UserController(use_case)

# Simulate HTTP requests
requests = [
    {"email": "john@example.com", "username": "john_doe"},
    {"email": "jane@example.com", "username": "jane_smith"},
    {"email": "", "username": "invalid_user"},  # This will fail
]

print("\n  User Creation Requests:")
for i, request in enumerate(requests, 1):
    print(f"    Request {i}: {request}")
    response = controller.create_user_endpoint(request)
    print(f"    Response: {response}")

# ===============================================================
# ARCHITECTURE PATTERNS SUMMARY
# ===============================================================
print("\n" + "=" * 75)
print("рҹҸҶ SOFTWARE ARCHITECTURE PATTERNS MASTERY")
print("=" * 75)

patterns = {
    "Microservices": "Independent services, fault isolation, technology diversity",
    "Event-Driven": "Loose coupling, real-time processing, reactive systems",
    "Clean Architecture": "Dependency inversion, testable, maintainable code"
}

print("\nрҹ“ҡ PATTERNS IMPLEMENTED:")
for pattern, description in patterns.items():
    print(f"вң… {pattern}: {description}")

print("\nрҹ’ј PRINCIPAL ENGINEER SKILLS:")
print("рҹҸҶ Design systems that scale to millions")
print("рҹҸҶ Choose appropriate architecture patterns")
print("рҹҸҶ Balance trade-offs and non-functional requirements")
print("рҹҸҶ Lead technical decision making")
print("рҹҸҶ Mentor teams on architectural best practices")

print("\nрҹҡҖ REAL-WORLD IMPACT:")
print("вң… Netflix: Microservices for global streaming")
print("вң… Uber: Event-driven for real-time matching")
print("вң… Google: Clean architecture for massive codebases")
print("вң… Amazon: Service-oriented for e-commerce scale")

print("\nрҹ’Ў ARCHITECTURE DECISION FACTORS:")
print("рҹ“Ҡ Team Size & Skills")
print("рҹ“Ҡ System Complexity")
print("рҹ“Ҡ Performance Requirements") 
print("рҹ“Ҡ Scalability Needs")
print("рҹ“Ҡ Maintenance & Evolution")
print("=" * 75)
````

## рҹҺҜ **Architecture Decision Framework**

When designing a system, ask:

1. **Scale Requirements**
   - <10K users вҶ’ Monolith
   - 10K-100K вҶ’ Modular monolith
   - >100K вҶ’ Microservices
   - Unpredictable вҶ’ Serverless

2. **Team Size**
   - <5 people вҶ’ Monolith
   - 5-20 вҶ’ Modular monolith
   - >20 вҶ’ Microservices

3. **Complexity**
   - Simple CRUD вҶ’ Layered
   - Complex business logic вҶ’ DDD/Hexagonal
   - Event-heavy вҶ’ Event-driven
   - Audit requirements вҶ’ Event sourcing

4. **Consistency Needs**
   - Strong вҶ’ ACID, CP
   - Eventual OK вҶ’ BASE, AP
   - Mix вҶ’ CQRS

5. **Deployment Frequency**
   - Monthly вҶ’ Monolith OK
   - Weekly вҶ’ Modular
   - Daily/Continuous вҶ’ Microservices

## рҹ’Ў **Interview Tips**

**Always justify your architecture choice:**

вқҢ Bad: "I'll use microservices"
вң… Good: "Given 1M users, multiple teams, and independent scaling needs, microservices make sense. We'll start with 5-10 services around core domains (users, products, orders), use API gateway for routing, and implement circuit breakers for resilience."

**Discuss trade-offs:**
- "Microservices give us scalability but add operational complexity"
- "Event sourcing provides audit trail but eventual consistency"
- "API Gateway centralizes auth but becomes SPOF (need HA)"

**Evolution path:**
"Start with monolith, extract microservices as needed"
Not: "Build 50 microservices on day 1"

**Remember: Architecture serves the business, not the other way around!**
